```python
from calendar import month
from cmath import cos, pi, sin
from math import degrees, radians
import cv2
from matplotlib import pyplot as plt
import numpy as np

from datetime import date,datetime


cv2.namedWindow("project")
while True:
    img=np.zeros((900,1300,3),np.uint8)

    now=datetime.now()
    day=int(now.strftime("%d"))
    mon=int(now.strftime("%m"))
    Year=int(now.strftime("%Y"))

    hours=int(now.strftime("%H"))
    mins=int(now.strftime("%M"))
    sec=int(now.strftime("%S"))
    usec=int(now.strftime("%f"))

    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img,f"Date : {day}-{mon}-{Year}",(25,50),font,1,(255,215,100),2)
    cv2.putText(img,f"Time : {hours}:{mins}:{sec}",(25,90),font,1,(255,215,0),2)
    r=250
    origin=[625,450]

    #draw big circle
    cv2.circle(img,origin,r,(255,117,0),3)

    #draw circles
    counter=3
    for n in range(60):
        theta = n*(pi/6)/5
        x1=int(r*cos(theta).real+origin[0])
        y1=int(r*sin(theta).real+origin[1])
        xt=int((r+20)*cos(theta).real+origin[0])
        yt=int((r+20)*sin(theta).real+origin[1])

        if (n%5) ==0:
            if xt<origin[0]:
                xt-=20
            if yt<origin[1]:
                yt-=30
            cv2.putText(img,f"{counter}",[xt,yt+20],font,1,(184,233,52),2)
            cv2.circle(img,[x1,y1],8,(0,255,255),2)
            counter+=1
            if counter>12:
                counter-=12
        else:
            cv2.circle(img,[x1,y1],4,(100,100,1000),2)

    #draw hours line
    if hours>12:
        hours-=12
    htheta=hours*(-pi/6)+(pi/2) - (mins/60)*(pi/6)

    xh=int((r-100)*cos(htheta).real+origin[0])
    yh=int((r-100)*-sin(htheta).real+origin[1])
    cv2.line(img,origin,[xh,yh],(145,145,0),6)

    #draw mins line
    mtheta=mins*(pi/6)/5+(pi/2)+(sec/60)*(pi/6)/5
    xm=int((r-40)*-cos(mtheta).real+origin[0])
    ym=int((r-40)*-sin(mtheta).real+origin[1])
    cv2.line(img,origin,[xm,ym],(145,0,145),3)
    
    #draw sec.s line
    stheta=sec*(pi/6)/5+(pi/2)+(usec/1000000)*(pi/6)/5
    xs=int((r-15)*-cos(stheta).real+origin[0])
    ys=int((r-15)*-sin(stheta).real+origin[1])
    cv2.line(img,origin,[xs,ys],(145,145,145),2)

    #draw centre circle
    cv2.circle(img,origin,10,(255,255,255),-1)
    
    

    cv2.imshow('project',img)
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
cv2.destroyAllWindows()



```
